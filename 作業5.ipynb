```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic_equation(a, b, c):
    x = np.linspace(1,100)  # 在指定範圍內生成一組 x 值
    y = a * x**2 + b * x + c  # 計算對應的 y 值

    # 繪製圖形
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Quadratic Equation')
    plt.grid(True)
    plt.show()

# 設定二元一次方程式的係數
a = -2
b = 4
c = 5

# 繪製二元一次方程式的圖形
plot_quadratic_equation(a, b, c)
```
