```python
import csv
import openpyxl

# 開啟CSV文件
with open('population.csv', 'r', encoding='utf-8') as csvfile:
    # 讀取CSV文件
    csvreader = csv.reader(csvfile)
    # 將資料存儲到列表中
    data = list(csvreader)

# 創建Excel檔案
workbook = openpyxl.Workbook()
# 創建一個工作表
worksheet = workbook.active
# 將CSV資料寫入工作表中
for row in data:
    worksheet.append(row)

# 保存Excel檔案
workbook.save('population.xlsx')
```

```python
import openpyxl

# 開啟Excel檔案
wb = openpyxl.load_workbook('excel-read.xlsx')

# 取得工作表名稱
sheet_names = wb.sheetnames

# 取得第一個工作表
ws = wb[sheet_names[0]]

# 讀取儲存格資料
for row in ws.iter_rows(min_row=2, values_only=True):
    year = row[0]
    population = row[1]
    print(year, population)
```

```python
import openpyxl
import matplotlib.pyplot as plt

# 加载Excel文件
wb = openpyxl.load_workbook('excel-read.xlsx', data_only=True)
sheet = wb['csv']

# 提取人口数和月份数据
population = []
months = []
for row in sheet.iter_rows(min_row=2, values_only=True):
    population.append(row[1])
    months.append(row[0])

# 绘制折线趋势图
plt.plot(population, months)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.title('Population Trend', fontsize=16)
plt.show()
```

```python
import openpyxl
import matplotlib.pyplot as plt

# 從Excel文檔中讀取數據
wb = openpyxl.load_workbook('excel-read.xlsx', data_only=True)
ws = wb['csv']

# 提取需要繪圖的資料
data = []
for row in ws.iter_rows(min_row=2, values_only=True):
    data.append(row[1])

# 使用matplotlib創建一個pie物件
plt.pie(data)

# 設置圓餅圖的標籤和顏色
plt.legend(labels=ws['B'][1:], loc="best")
plt.title("Population by Year")

# 顯示圓餅圖
plt.show()
```
