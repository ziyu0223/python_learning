```python
from typing_extensions import Counter
# 1～1000數字中找出同時無法被2、3、5、7整除的數並列出
# 上述共有幾個


count = 0   
for i in range(1,1001):   
    if i%2!=0 and i%3!=0 and i%5!=0 and i%7!=0:   
        print(i)   
        count += 1 

print("共有", count, "個數字")
```
