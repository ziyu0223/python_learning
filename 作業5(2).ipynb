```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic_equation(a, b, c):
    x = np.linspace(1, 100)  # 在指定范围内生成一组 x 值
    y = a * x**2 + b * x + c  # 计算对应的 y 值

    # 绘制图形
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Quadratic Equation')
    plt.grid(True)
    plt.show()

# 设置二次方程的系数
a = -3
b = 4
c = 7

# 计算方程的根
roots = np.roots([a, b, c])

# 标出根
plt.scatter(roots, [0, 0], color='red', label='Roots')

# 生成一组 x 值
x = np.linspace(1, 100)

# 计算对应的 y 值
y = a * x**2 + b * x + c

# 计算最大和最小值的位置
max_index = np.argmax(y)
min_index = np.argmin(y)

# 找到最大和最小值的坐标
max_coord = (x[max_index], y[max_index])
min_coord = (x[min_index], y[min_index])

# 标出最大和最小值的坐标
plt.scatter(*max_coord, color='red', label='Max')
plt.scatter(*min_coord, color='blue', label='Min')

# 添加坐标值的标注
plt.annotate(f'Max: {max_coord}', max_coord, xytext=(20, -30),
             textcoords='offset points', arrowprops=dict(arrowstyle='->', color='red'))

plt.annotate(f'Min: {min_coord}', min_coord, xytext=(20, 20),
             textcoords='offset points', arrowprops=dict(arrowstyle='->', color='blue'))

# 绘制二次方程的图形
plot_quadratic_equation(a, b, c)
```
